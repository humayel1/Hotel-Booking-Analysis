#### Chart - 4

```python
# Chart - 4 visualization code
# The revenue of the hotels

# Read the hotel bookings dataset
data = pd.read_csv("/content/drive/MyDrive/Hotel Bookings.csv")


# Calculate the revenue for each booking
data['revenue'] = data['adr'] * (data['stays_in_weekend_nights'] + data['stays_in_week_nights'])

# Group the data by hotel and calculate the total revenue
revenue_by_hotel = data.groupby('hotel')['revenue'].sum()

# Create a list of hotels and their revenue values
hotels = revenue_by_hotel.index
revenue_values = revenue_by_hotel.values

# Set the positions and width for the bars
positions = range(len(hotels))
width = 0.35

# Define custom colors for the bars
colors = ['blue', 'green']

# Create the grouped bar plot with custom colors
fig, ax = plt.subplots()
rects = ax.bar(positions, revenue_values, width, color=colors, label='Revenue')

# Set the x-axis labels and title
ax.set_xlabel('Hotel')
ax.set_ylabel('Revenue')
ax.set_title('Revenue of the Hotels')

# Set the x-axis ticks and labels
ax.set_xticks(positions)
ax.set_xticklabels(hotels)

# Add the revenue values on top of each bar
for rect in rects:
    height = rect.get_height()
    ax.annotate(f'{height:.2f}', xy=(rect.get_x() + rect.get_width() / 2, height), xytext=(0, 3),
                textcoords="offset points", ha='center', va='bottom')

plt.show()
```

##### 1. Why did you pick the specific chart?

* Categorical Data: The hotel bookings dataset contains categorical data, specifically the different hotels. A grouped bar plot is well-suited for visualizing categorical data as it clearly distinguishes between different categories and displays their corresponding values.
